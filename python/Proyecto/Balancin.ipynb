### <p class="ex2">     El balancín consta de dos contrapesos, uno de los cuales es capaz de ejercer una fuerza estabilizadora, gracias a la hélice y motor que lleva montados. Este sistema es básicamente un péndulo, con la adición de que puede mantenerse en una posición diferente a la horizontal.<br>          Para medir la posición en la que se encuentra la barra transversal, es usado un potenciómetro (más adelante se caracterizará). </p> <hr> <p class="ex1">Modelo Matemático</p> <p class="ex2">     La ecuación dinámica del sistema es:</p><br>     <center>$l_1(F_e - F_g\:sin\:\theta - \beta_1\omega) + l_2(F_e\:sin\:\theta-\beta_2\omega) = I\alpha$</center> <p class="ex2">En donde:</p> <ul class="a">     <li><p class="ex2">$l_1:=\ $ Distancia desde la masa izquierda hasta el centro</p></li>     <li><p class="ex2">$l_2:=\ $ Distancia desde la masa derecha hasta el centro</p></li>     <li><p class="ex2">$W_m:=\ $ Fuerza gravitatoria relativa al conjunto motor-hélice</p></li>     <li><p class="ex2">$F_e:=\ $ Fuerza de empuje generada por la hélice</p></li>     <li><p class="ex2">$W_c:=\ $ Fuerza gravitatoria relativa al contrapeso</p></li>     <li><p class="ex2">$\beta_1:=\ $ Coeficiente de rozamiento relativo al conjunto motor-hélice</p></li>     <li><p class="ex2">$\beta_2:=\ $ Coeficiente de rozamiento relativo al contrapeso</p></li>     <li><p class="ex2">$I:=\ $ Momento de inercia de la sección móvil del montaje</p></li>     <li><p class="ex2">$\alpha:=\ $ Aceleración angular</p></li> </ul> <p class="ex2">     El sistema se limitará a una zona de operación en que se cumple $sin\:\theta \approx\theta$. Entonces, la siguiente aproximación es correcta:<br><br> </p> <center>$l_1(F_e - W_m\:\theta - \beta_1\omega) + l_2(W_c\:\theta-\beta_2\omega) = I\alpha$</center><br> <p class="ex2">     Reescribiendo los términos como diferenciales:<br><br> </p> <center>$l_1(F_e - W_m\:\theta - \beta_1\frac{d\theta}{dt}) + l_2(W_c\:\theta-\beta_2\frac{d\theta}{dt}) = I\frac{d^2\theta}{dt^2}$</center><br> <p class="ex2">     Para que sea una EDO se debe hacer un cambio de variables, con el fin de que todas las ecuaciones diferenciales resultantes sean de primer orden.<br><br> </p>

### $$y = F_e$$<br> \[ \left\{   \begin{array}{l l}     x = \frac{d\theta}{dt}\\ \\     \frac{dx}{dt} = \frac{l_1(y - W_m\:\theta - \beta_1x) + l_2(W_c\:\theta-\beta_2x)}{I}   \end{array} \right.\] <br><br> <p class="ex2">Ahora, si se tiene en cuenta el valor numérico de las constantes usadas:<br><br></p> $$l_1 = 0.3\:m$$<br> $$l_2 = 0.25\:m$$<br> $$W_m = m_m\:g = (9.81\:m/s^2)(0.150\:kg) = 1.4715\:N$$<br> $$W_c = m_c\:g = (9.81\:m/s^2)(0.350\:kg) = 3.4335\:N$$<br> $$\beta_1 = \beta_2 = 0.0001\frac{N\cdot m}{rad/s}$$<br> $$I = 0.0765375\:kg\cdot m^2$$<br><br> <p class="ex2">El sistema resultante es:</p> \[ \left\{   \begin{array}{l l}     x = \frac{d\theta}{dt}\\ \\     \frac{dx}{dt} = \frac{l_1(y - W_m\:\theta - \beta_1x) + l_2(W_c\:\theta-\beta_2x)}{I}   \end{array} \right.\]
